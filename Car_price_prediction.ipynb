```python
# inorder to collect only the brand name , we use the lambda function
df['Brand']=df['Name'].apply(lambda x:x.split()[0])
df
```

```python
df['Brand'].value_counts()
```

```python
#  since the onehot encoding also has some drawbacks, here we use Get dummies.
# get dummies work as same as onehot encoding,in which the first column is droped
# and from the values of the remaining columns the object is identified.
# get dummies is an encoding technique that comes under pandas.
# all other encoding techniques are in ML
```

```python
# encoding the necessayr columns,along with dropping first column
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],dtype=int,drop_first=True)
df1
```

```python
# Joining the two data frames, df and df1
dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
dfe.columns
```

```python
# dropping unnecessary columns
dfe.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Brand'],axis=1,inplace=True)
dfe
```

```python
dfe.columns
```

```python
# for matching with testing data, removing the fuel-electric
dfe.drop(['Fuel_Type_Electric'],axis=1,inplace=True)
dfe
```

```python
dfe.columns
```

```python
# removing the string in the object values & replacing it with nonespace
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe['Power']=dfe['Power'].str.replace('bhp','')
# replacing the null values in these columns with zero
dfe['Engine']=dfe['Engine'].str.replace('null','0')
dfe['Power']=dfe['Power'].str.replace('null','0')
dfe['Mileage']=dfe['Mileage'].str.replace('null','0')
dfe
```

```python
dfe.dtypes
```

```python
# datatype conversion of these columns
dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Engine']=dfe['Engine'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
dfe
```

```python
dfe.dtypes
```

```python
dfe.loc[dfe.Engine==0,'Engine']=np.NaN
dfe.loc[dfe.Power==0,'Power']=np.NaN
dfe.loc[dfe.Mileage==0,'Mileage']=np.NaN
```

```python
dfe.isna().sum()
```

```python
dfe['Mileage']=dfe['Mileage'].fillna(dfe['Mileage'].mean())
dfe['Engine']=dfe['Engine'].fillna(dfe['Engine'].mean())
dfe['Power']=dfe['Power'].fillna(dfe['Power'].mean())
dfe['Seats']=dfe['Seats'].fillna(dfe['Seats'].mode()[0])
dfe.isna().sum()
```

```python
dfe.drop(['Brand_Audi'],axis=1,inplace=True)
dfe.drop(['Brand_Force'],axis=1,inplace=True)
dfe.drop(['Brand_Lamborghini'],axis=1,inplace=True)
dfe.drop(['Brand_Smart'],axis=1,inplace=True)
```

```python
# separating input and output features
x=dfe.drop(['Price'],axis=1)
x
```

```python
y=dfe['Price']
y
```
